```python
categories=df['main_category'].value_counts()
df_successful = df[(df["state"] == "successful")]["main_category"]
successful_cat =df_successful.value_counts()
success=np.zeros(len(categories))
for i in range(len(categories)):
    success[i]=successful_cat[categories.index.tolist()[i]] 
per_successful = [(success[i]/categories[i])*100 for i in range(len(categories))]

df_failed = df[(df["state"] == "failed")]["main_category"]
failed_cat =df_failed.value_counts()
failed=np.zeros(len(categories))
for i in range(len(categories)):
    success[i]=failed_cat[categories.index.tolist()[i]] 
per_failed = [(failed_cat[i]/categories[i])*100 for i in range(len(categories))]

df_canceled = df[(df["state"] == "canceled")]["main_category"]
canceled_cat =df_canceled.value_counts()
canceled=np.zeros(len(categories))
for i in range(len(categories)):
    canceled[i]=canceled_cat[categories.index.tolist()[i]] 
per_canceled = [(canceled_cat[i]/categories[i])*100 for i in range(len(categories))]

percentile_states=[per_successful, per_failed, per_canceled]
percentile_states

print("Categories = ")
print(categories)
print("Successful")
print(success)
print("Failed")
print(failed)
print("Canceled")
print(canceled)

for i in range(len(percentile_states)):
    print(percentile_states[i])
    print("***********************")

cols = categories.index.tolist()
rows = ['successful','failed','canceled']
values = np.arange(0, 100, 5)
index = np.arange(len(cols)) + 0.3

colors = plt.cm.BuPu(np.linspace(0, 0.5, len(rows)))
y_offset = np.zeros(len(cols))

# Plot bars and create text labels for the table
cell_text = []
for row in range(len(rows)):
    plt.bar(index, percentile_states[row], bottom=y_offset, color=colors[row])
    y_offset = y_offset + percentile_states[row]
    cell_text.append(['%1.1f' % (x / 1000.0) for x in y_offset])
    
# Reverse colors and text labels to display the last value at the top.
colors = colors[::-1]
cell_text.reverse()

# Add a table at the bottom of the axes
the_table = plt.table(cellText=cell_text,
                      rowLabels=rows,
                      rowColours=colors,
                      colLabels=cols,
                      loc='bottom')

# Adjust layout to make room for the table:
plt.subplots_adjust(left=0.2, bottom=0.2)

plt.xticks([])
plt.title('Loss by Disaster')

plt.show()
```
